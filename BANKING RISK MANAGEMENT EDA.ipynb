# BIVARIATE ANALYSIS

```python
predictors = ["pdays", "campaign", "duration", "marital", "education", "loan", "contact", 
              "previous", "poutcome", "job", "default", "balance", "housing", "day"]

for i, predictor in enumerate(df[predictors].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue="age")
    plt.title(f'Count plot of {predictor} by Age')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

# UNIVARIATE ANALYSIS

```python
predictors = ["pdays", "campaign", "duration", "marital", "education", "loan", "contact", 
              "previous", "poutcome", "job", "default", "balance", "housing", "day"]

for i, predictor in enumerate(df[predictors].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
    plt.title(f'Count plot of {predictor}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Hisplt of value counts for different jobs

for col in categorical_cols:
    if col == "job":
        plt.figure(figsize=(8, 4))
        sns.countplot(data=df, x=col)
        plt.title("Histogram of job count")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

# Numerical Columns 

```python
numerical_cols =["balance","age"]

#Univariate analysis and visuaization

plt.figure(figsize=(10, 6))

for i, col in enumerate(numerical_cols):
    plt.subplot(2, 1, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
import math
numerical_cols = ["pdays", "campaign", "duration", "balance", "previous", "day"]

n = len(numerical_cols)
cols = 2  # number of plots per row
rows = math.ceil(n / cols)

plt.figure(figsize=(14, 5 * rows))

for i, col in enumerate(numerical_cols):
    plt.subplot(rows, cols, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

## HEATMAPS

```python
# Select only numeric columns
numerical_cols = ["pdays", "campaign", "duration", "balance", "previous", "day", "age"]
numerical_df = df[numerical_cols]

# Compute correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.show()
```

# boxplot

```python
plt.figure(figsize=(14, 8))
for i, col in enumerate(numerical_cols):
    plt.subplot(math.ceil(len(numerical_cols)/2), 2, i + 1)
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```
